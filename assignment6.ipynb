```python
# q.3
class MultiplicationTable:
    def table(self, n):
        for i in range(1, 11):
            print(n, "x", i, "=", n * i)

obj = MultiplicationTable()

obj.table(5)
```

```python
# q.4
class FactorialCalculator:
    def factorial(self, n):
        fact = 1

        for i in range(1, n + 1):
            fact = fact * i
        print("Factorial =", fact)
obj = FactorialCalculator()
obj.factorial(4)
```

```python
# q.5 
class PrimeNumberChecker:
    def check(self):
        n = int(input("Enter a number: "))
        count = 0
        for i in range(1, n + 1):
            if n % i == 0:
                count += 1
        if count == 2:
            print(n, "is Prime")
        else:
            print(n, "is Not Prime")
obj = PrimeNumberChecker()
obj.check()
```

```python
# Q.6
class Student:
    def __init__(self, name, roll_no, course):
        self.name = name
        self.roll_no = roll_no
        self.course = course

    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.roll_no)
        print("Course:", self.course)

s1 = Student("Harsh", 12, "Python")
s1.display()
```

```python
#Q.10
class Vehicle:
    def vehicle(self):
        print("This is a Vehicle")

class Car(Vehicle):
    def car(self):
        print("This is a Car")

class ven(Car):
    def sports(self):
        print("Ferrari is a Sports Car")
s = ven()
s.vehicle()
s.car()
s.sports()
```

```python
#Q.11
class BankAccount:
    def __init__(self):
        self.balance = 10000

    def deposit(self):
        self.balance += 5000
        print("Balance after Deposit =", self.balance)

    def withdraw(self):
        self.balance -= 2000
        print("Balance after Withdraw =", self.balance)

    def check_balance(self):
        print("Current Balance =", self.balance)

b = BankAccount()

b.deposit()
b.withdraw()
b.check_balance()
```
