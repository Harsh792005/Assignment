##### 1. Write a Python script that accepts a number from the command line and checks whether it is prime. Explain how command-line arguments are accessed. 

```python
def check(numbers):
    small = numbers[0]
    big = numbers[0]
    for i in numbers:
        if i < small:
            small = i

        if i > big:
            big = i

    print("Smallest =", small)
    print("Largest =", big)

list1 = [10, 5, 20, 2]
check(list1)
```
