```python
class Calculator:
    def add(self, a, b):
        print("Addition =", a + b)

    def subtract(self, a, b):
        print("Subtraction =", a - b)

    def multiply(self, a, b):
        print("Multiplication =", a * b)

    def divide(self, a, b):
        print("Division =", a / b)

# User input
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

c = Calculator()
c.add(x, y)
c.subtract(x, y)
c.multiply(x, y)
c.divide(x, y)
```
