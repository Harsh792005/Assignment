##### Q4. Write a Python program to perform various operations on Python collections:
List operations (append(), remove(), slicing)
Tuple indexing
Set operations (union, intersection)
Dictionary operations (keys(), values(), items())

```python
#LIST OPERATIONS
numbers = [10, 20, 30, 40]
lst.append(50)
print(numbers)

# remove()
numbers.remove(20)
print("remove():", numbers)

# slicing
print("Slicing:", numbers[1:4])


#TUPLE INDEXING

data = ("Harsh", 21, "CSE")
print(data[1])

# SET OPERATIONS
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# union
print(set1.union(set2))

# intersection
print(set1.intersection(set2))


#  DICTIONARY OPERATIONS 
student = {
    "name": "Harsh",
    "branch": "CSE",
    "roll_no": 21
}
print(student)

# keys()
print("Keys:", student.keys())

# values()
print("Values:", student.values())

# items()
print("Items:", student.items())
```

##### Q5.Develop a mini Student Management System using Python data types.
The program should:
a. Store student details using Dictionary
b. Store subject marks using List
c. Calculate total and average marks
d. Display the output in proper format  very easy code

```python
# Student details using Dictionary
student = {
    "name": "Harsh",
    "roll_no": 21,
    "branch": "CSE"
}
marks = [80, 75, 90]
total = sum(marks)
average = total / len(marks)
# Display output
print("Student Details")
print("Name :", student["name"])
print("Roll No :", student["roll_no"])
print("Branch :", student["branch"])
print("\nMarks :", marks)
print("Total Marks :", total)
print("Average Marks :", average)
```
